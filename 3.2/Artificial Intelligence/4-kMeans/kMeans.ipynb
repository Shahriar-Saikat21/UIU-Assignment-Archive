```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Load samples from "jain_feats.txt" 
X = np.loadtxt("jain_feats.txt")

# Load initial centroids from "jain_centers.txt" 
centroid_old = np.loadtxt("jain_centers.txt")
```

```python
# Initialize centroid_new with zeros
centroid_new = np.zeros_like(centroid_old)
```

```python
# Initialize label array
label = np.zeros(X.shape[0], dtype=int)
```

```python
# Initial scatter plot
plt.scatter(X[:, 0], X[:, 1], c='blue', marker='.', label='Samples')
plt.scatter(centroid_old[:, 0], centroid_old[:, 1], c='red', marker='x', label='Centroids')
plt.title('Initial Scatter Plot')
plt.legend()
plt.show()
```

```python
# Number of iterations
iterations = 100
```

```python
for e in range(iterations):
    # Assign points to centroids(Eucledian distance from given two centroids)
    for i in range(X.shape[0]):
        dist = np.linalg.norm(X[i, :] - centroid_old, axis=1)
        label[i] = np.argmin(dist)
    # Update Centroids
    for j in range(centroid_new.shape[0]):
        centroid_new[j, :] = np.mean(X[label == j], axis=0)
    
    # Stop condition check
    if np.array_equal(centroid_new, centroid_old):
        break
    else:
        centroid_old = centroid_new.copy()
```

```python
# Plot the data points and final centroids
plt.scatter(X[:, 0], X[:, 1], c=label, marker='.', cmap = 'rainbow',label='Clusters')
plt.scatter(centroid_old[:, 0], centroid_old[:, 1], c='black', marker='x', label='Final Centroids')
plt.title('Final Scatter Plot with Clusters and Centroids')
plt.legend()
plt.show()
```
